```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#df = pd.read_csv("College.csv", index_col='Unnamed: 0')
#df.index.name = 'Name'

df = pd.read_csv("College.csv")
df.head()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14,6))
df.plot.scatter(y="Apps", x="Accept", ax=ax[0])
df.plot.scatter(y="Apps", x="Enroll", ax=ax[1])
df.plot.scatter(y="Accept", x="Enroll", ax=ax[:,0])
```

```python
plt.subplot(2,2,2)
df.plot.scatter(y="Apps", x="Enroll",alpha=0.5)
plt.subplot(2,2,3)
df.plot.scatter(y="Apps", x="Expend",alpha=0.5)
plt.subplot(2,2,4)
df.plot.scatter(y="Apps", x="Grad.Rate",alpha=0.5)
fig.subplots_adjust(wspace=1,hspace=1) # To add space between subplots
```

```python
fig = plt.figure()
plt.subplot(1,2,1)
df[df["Private"]=="Yes"]["Outstate"].plot.box(title="Private");
plt.subplot(1,2,2)
df[df["Private"]=="No"]["Outstate"].plot.box(title="Public");
fig.subplots_adjust(wspace=1,hspace=1) # To add space between subplots
```

```python
# make histogram by specified bins 
fig = plt.figure()
plt.subplot(2,2,1)
df['Apps'].hist(bins=10, alpha=0.5)
plt.subplot(2,2,2)
df['Apps'].hist(bins=20)
plt.subplot(2,2,3)
df['Apps'].hist(bins=30)
plt.subplot(2,2,4)
df['Apps'].hist(bins=40)
```

```python
# do it with seaborn
#f, axs = plt.subplots(1, 3, sharey=True)
f, axs = plt.subplots(2, 2)
sns.histplot(df, x="Apps", binwidth=5000, ax=axs[0,0])
sns.histplot(df, x="Apps", binwidth=1000, ax=axs[0,1])
sns.histplot(df, x="Apps", binwidth=500, ax=axs[1,0])
sns.histplot(df, x="Apps", binwidth=100, ax=axs[1,1])
f.tight_layout()
```

```python
# cut data into bins and make histogram
fig = plt.figure()
applications = pd.cut(df['Apps'], 3, labels=['Low', 'Medium', 'High'])
accepts = pd.cut(df['Accept'], 4, labels=['Very low', 'Low', 'High', 'Very high'])

plt.subplot(121)
applications.value_counts().plot(kind='bar', title = 'Applications');
plt.subplot(122)

accepts.value_counts().plot(kind='bar', title = 'Accept')
fig.subplots_adjust(hspace=1) # To add space between subplots

applications.value_counts()
```
